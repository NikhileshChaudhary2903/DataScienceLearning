```python
import numpy as np
import pandas as pd

# For visualization
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool

from bokeh.palettes import Spectral5
from bokeh.transform import factor_cmap
import pickle
```

```python
df=pd.read_csv('population/ca.1990_2016.19ages.txt',header=None,names=["oneval"])
```

```python
## 1990CA06001011010000004894
for index, row in df.iterrows():
    v=str(row['oneval'])
    year=int(v[0:4])
    state=str(v[4:6])
    state_fips_code=int(v[6:8])
    county_fips_code=int(v[8:11])
    registry=int(v[11:13])
    race=int(v[13:14])
    origin=int(v[14:15])
    sex=int(v[15:16])
    age=int(v[16:18])
    population=int(v[18:])


    df.at[index,'year'] = year
    df.at[index,'State_postal_abbreviation'] = state
    df.at[index,'State_FIPS_code'] = state_fips_code
    df.at[index,'County_FIPS_code'] = county_fips_code
    df.at[index,'Registry'] = registry
    df.at[index,'Race'] = race
    df.at[index,'Origin'] = origin
    df.at[index,'Sex'] = sex
    df.at[index,'Age'] = age
    df.at[index,'Population']= population
```

```python
df.drop(['oneval'], axis=1, inplace=True)
```

```python
df.columns
```

```python
## Filter only San Francisco
df=df[df['County_FIPS_code']==75]
```

```python
series=df.groupby("year")['Population'].sum()
df_total_pop_per_year=pd.DataFrame({'year':series.index, 'Population':series.values})
```
